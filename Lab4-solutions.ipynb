```python
n = 5
qr = qlib.QuantumRegister(n, 'q')
circ = qlib.QuantumCircuit(qr)
def cblock(circuit, qa, qb):
    circuit.h(qa)
    circuit.cx(qa, qb)
    
cblock(circ, qr[0], qr[2])
circ.barrier()
cblock(circ, qr[1], qr[2])

circ.draw(output='mpl')
```

```python
qr = qlib.QuantumRegister(1, 'q')
circ = qlib.QuantumCircuit(qr)

for i in range(8):
    if i == 3:
        circ.barrier()
    if i < 3:
        circ.x(qr[0])
        circ.z(qr[0])
    else:
        circ.h(qr[0])
        circ.y(qr[0])

circ.draw(output='mpl')
```

## Άσκηση 4.3

```python
def bell(circuit,register,position):
    circuit.h(register[position])
    circuit.cx(register[position], register[position+1])
    
qr = qlib.QuantumRegister(2, 'q')
circ = qlib.QuantumCircuit(qr)

bell(circ, qr, 0)

circ.draw(output='mpl')
```

```python
def bellij(circuit,register,p1, p2):
    circuit.h(register[p1])
    circuit.cx(register[p1], register[p2])
    circuit.barrier()
    
qr = qlib.QuantumRegister(3, 'q')
circ = qlib.QuantumCircuit(qr)

bellij(circ, qr, 0, 1)
bellij(circ, qr, 0, 2)

circ.draw(output='mpl')
```

```python
def cphase(circuit,register,p1,p2,l):
    circuit.cu1(l, register[p1], register[p2])
    circuit.barrier()
    
qr = qlib.QuantumRegister(3, 'q')
circ = qlib.QuantumCircuit(qr)

circ.x(qr[0])
cphase(circ, qr, 0, 1, 3.14)

circ.h(qr[1])
cphase(circ, qr, 1, 2, 3.14/2)

circ.draw(output='mpl')
```

```python
def cphase2(circuit,qb1,qb2,l):
    circuit.cu1(l, qb1, qb2)
    circuit.barrier()
    
qr = qlib.QuantumRegister(3, 'q')
circ = qlib.QuantumCircuit(qr)

circ.x(qr[0])
cphase2(circ, qr[0], qr[1], 3.14)

circ.h(qr[1])
cphase2(circ, qr[1], qr[2], 3.14/2)

circ.draw(output='mpl')
```

```python
def xc(circuit,qa,qb):
        circuit.h(qa)
        circuit.h(qb)
        circuit.cx(qa,qb)
        circuit.h(qa)
        circuit.h(qb)

        
qr = qlib.QuantumRegister(2, 'q')
circ = qlib.QuantumCircuit(qr)

circ.x(qr[0])
circ.barrier()
xc(circ, qr[0], qr[1])

circ.draw(output='mpl')
```

## Άσκηση 4.4

```python
qr = qlib.QuantumRegister(3, 'q')
circ = qlib.QuantumCircuit(qr)


bell(circ,qr,1)
circ.barrier()
xc(circ,qr[0],qr[1])
circ.h(qr[0])

circ.draw(output='mpl')
```

```python
n = 7
qr = qlib.QuantumRegister(n, 'q')
circ = qlib.QuantumCircuit(qr)

for i in range(n-1):
    bellij(circ, qr, i, (2*i+1)%n)
    
circ.draw(output='mpl')
```

```python
n = 6
qr = qlib.QuantumRegister(n, 'q')
circ = qlib.QuantumCircuit(qr)

for i in range(1,n):
    cphase(circ, qr, 0, i, 3.14/i)
    
circ.draw(output='mpl')
```
