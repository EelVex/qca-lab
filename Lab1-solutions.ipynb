```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister 
from qiskit import BasicAer, execute
```

## Άσκηση 1.1

```python
qr = QuantumRegister(2,'x')
qc = QuantumCircuit(qr)
qc.x(qr[1])
qc.draw(output='mpl')
```

```python
qr = QuantumRegister(1,'x')
cr = ClassicalRegister(1,'y')
qc = QuantumCircuit(cr,qr)
qc.x(qr[0])
qc.measure(qr[0],cr[0])
qc.draw(output='mpl')
```

## Άσκηση 1.2

```python
# Αν δεν το έχουμε συμπεριλάβει στην αρχή, πρέπει να φορτώσουμε τα 'BasicAer' και 'execute':
from qiskit import BasicAer, execute

from qiskit.tools.visualization import plot_histogram

engine = BasicAer.get_backend('qasm_simulator')
```

```python
qr = QuantumRegister(1,'x')
cr = ClassicalRegister(1,'y')
qc = QuantumCircuit(cr,qr)
qc.measure(qr[0],cr[0])
qc.draw(output='mpl')
```

```python
result = execute(qc, engine).result()
```

```python
plot_histogram(result.get_counts(qc))
```

```python
qr = QuantumRegister(2,'x')
cr = ClassicalRegister(2,'y')
qc = QuantumCircuit(cr,qr)
qc.x(qr[0])
qc.measure(qr,cr)
qc.draw(output='mpl')
```

```python
result = execute(qc, engine).result()
```

```python
plot_histogram(result.get_counts(qc))
```

## Άσκηση 1.5

```python
qbits = QuantumRegister(1,'q')
cbits = ClassicalRegister(1,'c')
circuit = QuantumCircuit(qbits, cbits)
circuit.y(qbits[0])
circuit.measure(qbits,cbits)
circuit.draw(output='mpl')
```

```python
result = execute(circuit, engine, shots=1000).result()
print(result.get_counts(circuit))
```

```python
circuit = QuantumCircuit(qbits, cbits)
circuit.x(qbits[0])
circuit.y(qbits[0])
circuit.measure(qbits,cbits)
circuit.draw(output='mpl')
```

```python
result = execute(circuit, engine, shots=1000).result()
print(result.get_counts(circuit))
```

Βλέπουμε οτι η δράση της Υ είναι ακριβώς η ίδια με τη δράση της Χ.

Στην πραγματικότητα, η Y αλλάζει και τη *φάση* του qubit αλλά η φάση δεν μπορεί να μετρηθεί. Σε επόμενο εργαστήριο θα δούμε με ποιο τρόπο η Υ διαφέρει από την Χ στην πράξη.
